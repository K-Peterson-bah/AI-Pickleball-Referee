```python
# import necessary libraries
from ultralytics import YOLO
import json
```

```python
#import API keys
with open("config.json") as config_file:
    config = json.load(config_file)
    Roboflow_key = config["ROBOFLOW_API_KEY"]
```

```python
#import demo dataset from roboflow
from roboflow import Roboflow
rf = Roboflow(api_key=Roboflow_key)
project = rf.workspace("aipickleballref").project("ai-pickleball-referee")
version = project.version(3)
dataset = version.download("yolov8")
```

# YoloV8

```python
#build model
yolov8 = YOLO("yolov8n.pt")
```

```python
#Train model
results = yolov8.train(data = "Ai-pickleball-referee-3/data.yaml", epochs=15, imgsz = 1280, batch=4)
```
